```python
# Загрузить таблицу из файла `Fishing.csv` и сохранить её в датафрейм `dat`. 
# Вывести на экран первые 8 строк загруженного датафрейма.

import pandas as pd
import numpy as np
dat = pd.read_csv('Fishing.csv')
```

```python
# Добавить, используя метод `.apply()`, столбец `log_income`, содержащий натуральный логарифм доходов респондентов.

dat['log_income'] = dat.loc[:, 'income'].apply(np.log, axis = 1)
```

```python
# Посчитать для каждого респондента абсолютное значение отклонения `price` от `pbeach`  и сохранить результат в столбец `pdiff`.

func_abs_deviation = lambda x: abs(x['price'] - x['pbeach'])
dat['pdiff'] = dat.loc[:, 'price':'pbeach'].apply(func_abs_deviation, axis=1)
```

```python
# Сгруппировать наблюдения в таблице по признаку тип рыбалки (`mode`) и вывести для каждого типа среднюю цену (`price`),
# которую респонденты заплатили за рыбалку.

print(dat.groupby('mode').agg(np.mean).loc[:, 'price'])
```

```python
# Сгруппировать наблюдения в таблице по признаку тип рыбалки (`mode`) и вывести для каждого типа
# разницу между медианным и средним значением цены (`price`), которую респонденты заплатили за рыбалку.

print(dat.groupby('mode').agg({'price': lambda x: x.median() - x.mean()}))
```

```python
# Сгруппировать наблюдения в таблице по признаку тип рыбалки (`mode`) и сохранить полученные
# датафреймы (один для каждого типа рыбалки) в отдельные csv-файлы.
# В итоге должно получиться четыре разных csv-файла.

for name, data in dat.groupby("mode"):
    data.to_csv('{file_name}.csv'.format(file_name = name))
```

```python
# Отсортировать строки в датафрейме в соответствии со значениями `income` в порядке убывания таким
# образом, чтобы результаты сортировки сохранились в исходном датафрейме.

dat.sort_values('income', ascending = False, inplace = True)
```
